```python
import os
import csv

#Initializing an empty dictionary and list
votes_count={}
percentage_vote=[]

csvpath=os.path.join(".","election_data.csv")
with open (csvpath, newline="") as csvfile:
    csvreader=csv.reader(csvfile, delimiter=",")
    header=next(csvreader)
    
    for row in csvreader:
        # If the candidate is in votes_count, add 1
        if row[2] in votes_count.keys():
            votes_count[row[2]]+=1
        # Else add the candidate to votes_count, set the value to 1
        else:
            votes_count[row[2]]=1  

#Calculating total votes
total_votes=sum(votes_count.values())

#another way to calculate total votes
#total_votes= votes_count["Khan"]+votes_count["Correy"]+votes_count["Li"]+votes_count["O'Tooley"]

candidates=tuple(votes_count.keys())
votes=tuple(votes_count.values())

#Calculating percentage of votes 
for vote in votes:
    percentage_vote.append(round(vote/total_votes*100,2))

#Formatting the percentage votes
formatted_percentage=["{}%".format(percentage) for percentage in percentage_vote]


#Finding the maximum votes from the dictionary
maximum_votes=max(votes_count.values())

#Finding the candidate from dictionary corresponding to maximum votes  
winner=[key for key, value in votes_count.items() if value==maximum_votes]

#winner can also be found by using max and zip funtion, however, the values must be placed first .
#winner=max(zip(votes_count.values(),votes_count.keys()))


#Final Result
print("Election Results")
print("-------------------------")
print("Total Votes:",total_votes)
print("-------------------------")
print(candidates[0],":", formatted_percentage[0], votes[0])
print(candidates[1],":", formatted_percentage[1], votes[1])
print(candidates[2],":", formatted_percentage[2], votes[2])
print(candidates[3],":", formatted_percentage[3], votes[3])
print("-------------------------")
#We can also print the result using zip 
print(list(zip(candidates,votes,formatted_percentage)))
print("-------------------------")
print("Winner:", winner)
print("-------------------------")

file_to_output = os.path.join(".", "election_analysis.txt")
with open(file_to_output, "w") as txt_file:
    txt_file.write("Election Results")
    txt_file.write("\n")
    txt_file.write("-------------------------")
    txt_file.write("\n")
    txt_file.write(f"Total Votes: {total_votes}")
    txt_file.write("\n")
    txt_file.write("-------------------------")
    txt_file.write("\n")
    txt_file.write(f"{candidates[0]}: {formatted_percentage[0]} ({votes[0]})")
    txt_file.write("\n")
    txt_file.write(f"{candidates[1]}: {formatted_percentage[1]} ({votes[1]})")
    txt_file.write("\n")
    txt_file.write(f"{candidates[2]}: {formatted_percentage[2]} ({votes[2]})")
    txt_file.write("\n")
    txt_file.write(f"{candidates[3]}: {formatted_percentage[3]} ({votes[3]})")
    txt_file.write("\n")
    txt_file.write("-------------------------")
    txt_file.write("\n")
    txt_file.write(f"Winner: {winner}")
    txt_file.write("\n")
    txt_file.write("-------------------------")
```
